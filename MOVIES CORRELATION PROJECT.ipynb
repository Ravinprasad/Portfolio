```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline

#Changing the plot style
matplotlib.rcParams['figure.figsize'] = (12,8) 

#Reading the file
df = pd.read_csv(r'/Users/ravinprasadramachandern/Downloads/movies.csv')
```

```python
#Looking at the data

df.head()
```

```python
#Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Looking at data types

df.dtypes
```

```python
#Standardizing the values

df = df.fillna(0)

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df.head()
```

```python
#Sorting the values in terms of Gross Revenue

df.sort_values(by=['gross'], inplace=True, ascending=False)
```

```python
#Dropping any duplicates

df['company'].drop_duplicates()
```

```python
#Building a scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross revenue')
plt.ylabel('Budget')
plt.show()
```

```python
#Building a regression plot to better visualize

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Setting up a correlation using the Pearson method

df.corr(numeric_only=True)
```

```python
#Setting up a correlation matrix with heatmap

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.show
```

```python
#Numerising all the datatypes

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
    
```

```python
df.head()
```

```python
#Setting up a correlationmatrix for all the numerized values

correlation_matrix = df_numerized.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.show
```

```python
#Viewing the correlation

df_numerized.corr()
```

```python
#Sorting the correlations into pairs

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
#Finding out the highest correlation for Gross Revenue

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
